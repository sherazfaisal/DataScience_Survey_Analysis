```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('F://Topic_Survey_Assignment.csv')
```

```python
df
```

```python
df = df.sort_values(["Very interested"], ascending = (False))
df['Very interested'] = df['Very interested'].apply(lambda x: x*100/2233)
df['Somewhat interested'] = df['Somewhat interested'].apply(lambda x: x*100/2233)
df['Not interested'] = df['Not interested'].apply(lambda x: x*100/2233)
df = df.reset_index(drop=True)
df
```

```python
df['Very interested'] = df['Very interested'].apply(lambda x: round(x,2))
df['Somewhat interested'] = df['Somewhat interested'].apply(lambda x: round(x,2))
df['Not interested'] = df['Not interested'].apply(lambda x: round(x,2))
df
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['Data Analysis / Statistics', 'Machine Learning', 'Data Visualization', 'Big Data (Spark / Hadoop)', 'Deep Learning', 'Data Journalism']

x = np.arange(len(labels))# the label locations
y = np.arange(100)
width = 0.25  # the width of the bars
#fig= plt.figure(figsize=(20,8))
fig, ax = plt.subplots(figsize=(20,8))
rects1 = ax.bar(x-width/3, df['Very interested'], width, label='Very interested', color = '#5cb85c')
rects2 = ax.bar(x+2*width/3, df['Somewhat interested'], width, label='Somewhat interested', color = '#5bc0de')
rects3 = ax.bar(x+5*width/3, df['Not interested'], width, label='Not interested', color = '#d9534f')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage', fontsize =18)
ax.set_title("Percentage of Respondent' interest in Data Science", fontsize =22)
ax.set_xticks(x)
ax.set_xticklabels(labels , fontsize =18)
ax.set_yticklabels(y*10,fontsize =18)
ax.legend(fontsize = 18)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2 , height),
                    xytext=(0, 2),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
